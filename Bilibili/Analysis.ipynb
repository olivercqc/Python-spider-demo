```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
import datetime
sns.set()
%matplotlib inline
```

```python
data = pd.read_csv("movie_data.csv")
```

```python
data
```

```python
timestamp = data["timestamp"]
local_time = [time.localtime(i) for i in timestamp]
update_time = [time.strftime("%Y-%m-%d %H:%M:%S", i) for i in local_time]
data["update_time"] = update_time
```
