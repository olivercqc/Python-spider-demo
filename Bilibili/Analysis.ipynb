```python
import pandas as pd
import time
```

```python
data = pd.read_csv("movie_data.csv", encoding="utf-8")
```

```python
data.head()
```

```python
timestamp = data["timestamp"]
local_time = [time.localtime(i) for i in timestamp]
update_time = [time.strftime("%Y-%m-%d %H:%M:%S", i) for i in local_time]
data["update_time"] = update_time
```

```python
data.head()
```

```python
data.to_csv("new_movie_data.csv", index=False, encoding='utf-8')
```

```python
data.drop(columns=["timestamp"]).to_excel("new_data.xlsx", index=False, encoding="utf-8")
```
